```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
df= pd.read_csv('C:\\Users\\user\\Downloads\\openml_phpZNNasq.csv', sep=",")
```

```python
df.head()
```

```python
df.drop(columns='animal', inplace=True)
df.head()
```

```python
y = df.pop('type')
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.4, random_state=188)
```

```python
rf_model = RandomForestClassifier(random_state=42,n_estimators=30)
```

```python
rf_model.fit(X_train, y_train)
```

```python
train_preds = rf_model.predict(X_train)
train_preds
```

```python
train_acc = accuracy_score(y_train, train_preds)
print(train_acc)
```

```python
test_preds = rf_model.predict(X_test)
```

```python
test_acc = accuracy_score(y_test, test_preds)
print(test_acc)
```
